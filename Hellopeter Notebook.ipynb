__Geolite API__

```python
import pandas as pd
from geolite2 import geolite2
```

```python
#Use IP Address to get location
def get_location(ip):
    try:
        x = geolite2.reader().get(ip)
    except ValueError:
        return {'Continent': 'None','Country': 'None','City': 'None','Latitude': 'None','Longitude': 'None','Postal Code': 'None','Province': 'None'}
    
    try:
        return {'Continent':x['continent']['names']['en'],'Country':x['country']['names']['en'],'City':x['city']['names']['en'],'Latitude':x['location']['latitude'],'Longitude':x['location']['longitude'],'Postal Code':x['postal']['code'],'Province':x['subdivisions'][0]['names']['en']} 
    except KeyError:
        
        return {'Continent': 'None','Country': 'None','City': 'None','Latitude': 'None','Longitude': 'None','Postal Code': 'None','Province': 'None'}
```

```python
List_df =[]

for i in range(len(df)):
   
    List_df.append(get_location(df.author_ip_address[i]))
```

```python
Loc =pd.DataFrame(List_df)
```
